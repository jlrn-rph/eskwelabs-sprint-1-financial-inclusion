# Women Financial Inclusion

```python
# Import libraries
import pandas as pd
```

```python
# Load and read the data
data = pd.read_csv("../data/micro_world.csv", encoding="ISO-8859-1")
data.head()
```

```python
# Filter to Philippine Data and reset the index
philippine_data = data[data["economy"] == "Philippines"].reset_index(drop=True)

# View first 5 rows
philippine_data.head()
```

```python
# Make a copy of the dataframe
ph_data = philippine_data.copy()
ph_data.head()
```

```python
# Create new column 'has_account' and set to False (0) by default
ph_data["has_account"] = 0

# Check if all columns have a value of 1 and set 'has_account' to True (1)
ph_data.loc[(ph_data["account_fin"] == 1) | (ph_data["account_mob"] == 1), "has_account"] = 1
ph_data.head()
```

```python
# Count the value of has_account column
# 1 means that it has either an account in financial institution or mobile
ph_data["has_account"].value_counts()
```

```python
# Count the value of female column
ph_data["female"].value_counts()
```
